```python
# Import necessary libraries
import networkx as nx
import matplotlib.pyplot as plt
from itertools import combinations

# Function to perform Girvan-Newman algorithm
def girvan_newman(G):
    """Implementation of the Girvan-Newman algorithm for community detection."""
    # Create a copy of the graph to avoid modifying the original
    G_copy = G.copy()
    
    # Continue until there are no edges left
    while G_copy.number_of_edges() > 0:
        # Find the edge with the highest betweenness centrality
        edge_betweenness = nx.edge_betweenness_centrality(G_copy)
        max_edge = max(edge_betweenness, key=edge_betweenness.get)
        
        # Remove the edge with the highest betweenness
        G_copy.remove_edge(*max_edge)
        
        # Check the connected components
        components = list(nx.connected_components(G_copy))
        
        # If more than one component is found, return them
        if len(components) > 1:
            return components
    
    return [list(G_copy.nodes)]  # Return remaining nodes as a single component

# Create a sample graph
G = nx.Graph()

# Adding edges to the graph (you can customize this part)
edges = [(1, 2), (1, 3), (2, 3), (2, 4), (3, 5), (4, 5), (4, 6), (5, 7)]
G.add_edges_from(edges)

# Plot the initial graph
plt.figure(figsize=(8, 6))
nx.draw(G, with_labels=True, node_color='lightblue', node_size=700, font_size=16)
plt.title("Original Graph")
plt.show()

# Run the Girvan-Newman algorithm
communities = girvan_newman(G)

# Display the detected communities
print("Detected communities:", communities)

# Visualize the resulting communities
color_map = []
for node in G.nodes:
    if node in communities[0]:
        color_map.append('red')
    elif node in communities[1]:
        color_map.append('blue')
    else:
        color_map.append('green')

plt.figure(figsize=(8, 6))
nx.draw(G, with_labels=True, node_color=color_map, node_size=700, font_size=16)
plt.title("Detected Communities")
plt.show()
```
